First we need pull up the Import packages and Visualizations packages.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Then we need add in the data files that we saved.


```python
df = pd.read_csv('starbucks_drinkMenu_expanded.csv', encoding="ISO-8859-1", low_memory=False)
```

```python
df.head()
```

The data types for the dataset:


```python
df.dtypes
```

```python
df.calories = df[['Beverage_category','Calories']]
df.calories.columns = ['DrinkMenu', 'Calories']
df.calories
```

```python
sns.countplot(y='DrinkMenu', data=df.calories)
```

```python
df_beverage = pd.DataFrame(df['Beverage_category'].value_counts())
df_beverage['BeverageCategory'] = df_beverage.index
df_beverage.columns = ['TotalNumber', 'BeverageCategory']
df_beverage = df_beverage.reset_index().drop('index', axis=1)
df_beverage.head()
```

Here is my first question: What kind of drink that in Starbucks menu has the highest Calories?


```python
df.calories.sort_values("Calories", ascending=False)
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y="Calories", data=df)
plt.xticks(rotation=40, ha='right')
plt.title("Beverage Calories")
plt.show()
```

As we can see White chocolate mocha and Java Chip are two beverage with the most calories.

Can you make a histogram?
The answer is yes. We still want to focus to the calories in the Starbucks drinks so I plot a historgram of calories data.

```python
calories = df["Calories"]
plt.hist(calories, bins=9, edgecolor = "yellow", color = "blue")
plt.title("Calories in Starbucks Menu")
plt.xlabel("kcal")
plt.ylabel("Count")
```

The histogram shows that Starbucks has a large number of drinks based on 200 calories, but there are also some low-calorie drinks provided.
